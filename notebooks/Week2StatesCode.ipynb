## Entity code for states based on the insitution that provides the data

This Notebook uses the week1 data analyzes and a dataframe obtain trough the Mexican Census:  [INEGI](https://www.inegi.org.mx/app/tabulados/interactivos/?pxq=Educacion_Educacion_01_72949576-02b2-48ba-8e9e-58e6412c1fd0).

The purpose of this notebook organizes the states based on the code of the entity so no matter which institutions data is manage the code is the final determinate to identefy the entity. 

```python
import pandas as pd
```

```python
pd.options.display.max_rows = None  # display all rows
pd.options.display.max_columns = None  # display all columsns
```

The INEGI data is open in a way were only the columns of interest are use.

```python
df = pd.read_excel('../data/Educacion_01.xlsx',skiprows = range(0, 3))
df.head()
```

The column that refer the state is rename into a insitution recognizable name.

```python
df.rename(columns = {'Entidad federativa': 'state_inegi'}, inplace=True)
```

Since the states names are the only data of interest the whole country data row is delete.

```python
df.query("state_inegi != 'Estados Unidos Mexicanos'", inplace=True)
```

The footer of the dataframe is delete to only have the subset of the names of the states of interest

```python
df = df.iloc[0:32,:].copy()
df.head()
```

The names are cleaned in a way were the accents are eliminated and the letter are in uppercase. 

```python
df['state_inegi'] = df['state_inegi'].str.upper().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df.head()
```

The data is sorted through the state name in an ascending way (A-Z).

```python
df.sort_values(by=['state_inegi'], ascending = True, inplace=True)
```

A new columns is created for other institution (CONACYT) since this institution uses other names for the states which is why this is done to manage the data that it provides data to our analyzes.

```python
df['state_conacyt']=df['state_inegi'].copy()
df
```

The names of the states are changed from how INEGI denominates each to how CONACYT does.

```python
df.at[9,'state_conacyt']='DISTRITO FEDERAL'
df.at[5,'state_conacyt']='COAHUILA'
df.at[15,'state_conacyt']='MEXICO'
df.at[16,'state_conacyt']='MICHOACAN'
df.at[30,'state_conacyt']='VERACRUZ'
df.at[22,'state_conacyt']='QUERETARO'
df.at[24,'state_conacyt']='SAN LUIS POTOSI'
df.at[31,'state_conacyt']='YUCATAN'
df.at[19,'state_conacyt']='NUEVO LEON'
```

The names of the states per institution is grouped in a subset dataframe.

```python
df = df[['state_inegi','state_conacyt']].copy()
df.head()
```

The data from week 1 analyzes is open but only using the code of entity (cve_ent) and the name of the state (state) which is on CONACYT denomination.

```python
dfWeek1 = pd.read_csv('../data/week1analyzesStates.csv',usecols=['cve_ent','state'])
dfWeek1.head()
```

The name of the column state is changed to make it visible that it's on CONACYT denomination.

```python
dfWeek1.rename(columns = {'state':'state_conacyt'},inplace=True)
```

Both the names of the states of each instituion and the week 1 analyzes is merged to obtain the cve_ent of the state regardless of which institution state name is being used.

```python
dfStates = pd.merge(dfWeek1,df,on=['state_conacyt'])
dfStates.head()
```

The dataframe is sorted by the cve_ent in an ascending way (0->inf) to have a better order in the data.

```python
dfStates.sort_values(by=['cve_ent'], ascending = True, inplace=True)
dfStates.head()
```

The organized dataframe is store in the data folder so future studys with data provided by INEGI or other organizations can be manage in a simple way.

```python
dfStates.to_csv('../data/week2_states_code_institution.csv',index=False)
```
